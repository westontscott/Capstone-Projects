# __Rollout Plan__

There are 40 different classrooms for the 6th grade class in question. Luckily, students were randomly placed into each classroom by a random seed when sign ups for the Academy were finalized before the start of the year. This is good because each classroom can be considered a sample of the population. Rollout will be best suited with only 10% of teachers implementing this in their classrooms. 100 students spread across 4 different classrooms will be enough to get a test, while also not causing such a noticeable 50/50 split that parents begin to be alarmed that a test is going on. 

The plan is to introduce a 20 min creative writing journal to the test students curriculum. 

# __Hypothesis__

__Null Hypothesis__: There will not be a difference in writing score average improvement: mu_prep = mu_control

__Alternate Hypothesis__: Writing score average improvement will be higher than the control: mu_prep > mu_control

# __Success Metric__

We're going to run a t-test with a p-value of 5%. Our success metric will be based on writing score improvement on the mid-year exam. 

# __Secondary Metric__

It will be important to see how this extra concentration on writing will affect student performance on other subjects. If there is evidence to show that decreased performance in Math and Reading outweighs the increase in Writing performance, then it would no longer make sense to pursue this test.
